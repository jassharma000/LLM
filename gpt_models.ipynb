Generating Text

```python
def generate_text(prompt):
    response = openai.completion.create(
        engine="text-davinci-002",
        prompt=prompt,
        max_tokens=10,
        temperature=0.7)
    return response.choices[0].text.strip()
```

```python
prompt="Once upon a time"
```

```python
generated_text=generate_text(prompt)
print(generated_text)
```
